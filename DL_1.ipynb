```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("HousingData.csv")
df.head(20)
```

```python
df.isnull().sum()
```

```python
print("Shape",df.shape)
```

```python
df.fillna(df.mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df.drop(['MEDV'],axis=1)
y=df['MEDV']
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=45) #after x_train,x_test
```

```python
#Normalization first always
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
x_train=mms.fit_transform(x_train)
x_test=mms.transform(x_test)
```

```python
!pip install tensorflow
```

```python
#model

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model=Sequential()

model.add(Dense(128,input_shape=x_train[0].shape,activation='relu',name='dense1')) #train[0].shape   input_shape
model.add(Dense(64,activation='relu',name='dense2'))
model.add(Dense(32,activation='relu',name='dense3'))
model.add(Dense(1,activation='linear',name='output_dense'))

model.compile(optimizer='adam',loss='mae',metrics=['mae','mse'])

model.summary()
```

```python
m=model.fit(x_train,y_train,epochs=300)
```

```python
loss_n,mse_n,mae_n=model.evaluate(x_test,y_test) #test
print("Mean squared_error on test data",mse_n)
print("Mean absolute erroe on test data",mae_n)
```

```python
y_pred=model.predict(x_test)
y_test=np.array(y_test).reshape(-1,1)
plt.plot(y_test,label='Actual')
plt.plot(y_pred,label='Predicted')
plt.legend()

```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test, y_pred)
print("R² Score:", r2)

#close to 1 means perfoeming well
```
